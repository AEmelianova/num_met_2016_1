### Метод Хойна
*Метод Хойна является двукратным методом Рунге-Кутты при параметре $\alpha=1/2$.
Суть метода состоит в следующей формуле*
$$\begin{cases}
y_{n+1}=y_n+\frac{h}{2}(f(x_n,y_n)+f(x_{n+1},y_n+hf(x_n,y_n))\\
y_0=y^0, n=0,1....N-1, x_n\in w_n
\end{cases}$$

*Метод требует два вычисления функции f(x,y) на каждом шаге*

*С помощью данного метода было решено дифференциальное уравнение $5y''+8y'+4y=xe^{-x}$ с относительной
точностью 0,01*

$$\begin{cases}
y_{n+1}=y_n+\frac{h}{6}(k_1+2k_2+2k_3+k_4)\\
y_0=y^0, n=0,1...N-1, x_n\in w_n
\end{cases}$$
$$k_1=f(x_n,y_n)\\
k2=f(x_n+\frac{h}{2}, y_n+\frac{hk_1}{2})\\

*C помощью этого метода также было решено уравнение $5y''+8y'+4y=xe^{-x}$ с относительной

```python
import numpy
import pylab
def metod(func, func2, L0, step,a,b, delta):#L0-нач данные, step-стартовый шаг, а в - края отрезка
    y1=L0[0]
    y2=L0[1]
    Y0=numpy.zeros([2,int((b-a)/step)], "f")#решение
    Y1=numpy.ones([2,int((b-a)/step)], "f")
    n=0
    Y0[0][0]=y1
    Y0[1][0]=y2
    x1=a
    x2=a
    #делаем первое вычисление для базовой сетки отдельно
    for i in range(1,int((b-a)/step)):
        x2=x2+step
        #Y0[0][i]=Y0[0][i-1]+(step/2)*(func2(Y0[1][i-1],args)+func2(Y0[1][i-1]+step*Y0[1][i-1],args+step*func2(Y0[1][i-1],args)))
        #Y0[1][i]=Y0[1][i-1]+(step/2)*(func(x1,y1,y2)+func(x2,y1+step*func(x1,y1,y2),y2+step*func(x1,y1,y2)))
        Y0[0][i]=Y0[0][i-1]+(step/2)*(2*Y0[1][i-1]+step*Y0[1][i-1])
        Y0[1][i]=Y0[1][i-1]+(step/2)*(func(x1,y1,y2)+func(x2,y1+step*func(x1,y1,y2),y2+step*func(x1,y1,y2)))
        x1=x1+step
        y1=Y0[0][i]
        y2=Y0[1][i]
        #args=Y0[:,i]
    #print(Y0)
    #делаем деление сетки на 2 на каждом шаге, пока не наберем точность
    while True:
        x1=a
        x2=a
        if numpy.linalg.norm(Y0-Y1)/numpy.linalg.norm(Y1)>delta:
            newstep=step/2
            for j in range(0,int((b-a)/step)):
                Y1[0][j]=Y0[0][j]
                Y1[1][j]=Y0[1][j]
            Y2=numpy.zeros([2,int((b-a)/newstep)], "f")
            for i in range(1,int((b-a)/newstep)):
                n=n+1
                x2=x2+newstep
                #Y2[0][i]=Y2[0][i-1]+(newstep/2)*(func2(Y2[1][i-1],args)+func2(Y2[1][i-1]+newstep*Y2[1][i-1],args+newstep*func2(Y2[1][i-1],args)))
                Y2[0][i]=Y2[0][i-1]+(newstep/2)*(2*Y2[1][i-1]+newstep*Y2[1][i-1])
                Y2[1][i]=Y2[1][i-1]+(newstep/2)*(func(x1,y1,y2)+func(x2,y1+newstep*func(x1,y1,y2),y2+newstep*func(x1,y1,y2)))
                x1=x1+newstep
                #y1=Y2[0][i]
                #y2=Y2[1][i]
        else:
            break
        #записываем в Y0 только те узлы, которые были в изначальной сетке
        for i in range(1,int((b-a)/newstep)):
            if i%(2**n)==0:
                Y0[0][i/(2**n)]=Y2[0][i]
                Y0[1][i/(2**n)]=Y2[1][i]
    print(n)
    return Y0
L0=[1,0]
#L1=[0,1]
a=0; b=2; step1=0.01; #args=(1,)
ans=metod(func1,func2,L0,step1,a,b,0.01)
Ylist=[]; Ylist1=[] 
Ysolve=[]; DLT=[]
Xlist=numpy.arange(a,b,step1)
for i in range(0,int((b-a)/step1)):
    Ylist.append(ans[0][i])
    Ylist1.append(ans[1][i])
    Ysolve.append(answer(Xlist[i]))
    DLT.append(abs((Ysolve[i]-Ylist[i])/Ysolve[i]))
#print(DLT)
pylab.subplot(2,1,1)
plt.plot(Xlist, DLT, 'r')
plt.xlabel("X")
plt.ylabel("Delta")
plt.grid()
pylab.subplot(2,1,2)
plt.plot(Ylist,Ylist1,'b')
plt.xlabel("Y")
plt.ylabel("Y'")
plt.grid()
plt.show()
```
